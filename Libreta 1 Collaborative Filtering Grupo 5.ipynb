# Collaborative Filtering Grupo 5 Libreta 1

```python
# Importaion de la libreeria para la sesion de spark
from pyspark.sql import SparkSession
```

```python
import pandas as pd
```

```python
spark = SparkSession.builder.appName('libreta1CFI').getOrCreate()
```

```python
# Lectura del documento .csv con los datos de las reviews
data = spark.read.csv('amz-phones-reviews.csv', inferSchema = True, header=True)
```

```python
# Impresion de la estructura
data.printSchema()
```

```python
# RAW Data
data.toPandas()
```

```python
# Importacion de la libreria para las funciones especificadas
from pyspark.ml.feature import StringIndexer
from pyspark.ml import Pipeline

# Se indexan los "asin" que son los codigos del producto y "name" que es el identificador de quien hizo la review
asin_indexer = StringIndexer(inputCol='asin', outputCol='asin_index')
name_indexer = StringIndexer(inputCol='name', outputCol='name_index')
```

```python
# Se corre el pipeline para sacar los valores indexados
pipeline = Pipeline(stages=[asin_indexer, name_indexer])
# Los resultados del pipeline son guardados en el dataframe a evaluar
all_reviews = pipeline.fit(data).transform(data)
```

```python
all_reviews.toPandas()
```

```python
# Mas adelante se hara una verificacion de los reviews verificados, aqui se filtran solo los que sean True
verified_reviews = all_reviews.filter(all_reviews['verified']==True)
```

```python
# Se describen los valores que tienen los datos que se trabajara
all_reviews.select('asin_index', 'name_index', 'rating').describe().show()
```

## All Reviews

```python
# mportancion de la libreria de Collaborative Filtering
from pyspark.ml.recommendation import ALS
# Creacion del modelo clasificatorio para los datos
# ALS = MINIMOS CUADRADOS ALTERNOS
als = ALS(maxIter=25, userCol="name_index", itemCol="asin_index", ratingCol="rating", 
          coldStartStrategy="drop", nonnegative=True)
```

```python
# DIvision de los datos en entrenamiento y testeo en una proporcion de 80:20
(train, test) = all_reviews.randomSplit([0.8, 0.2])
```

```python
# Se introduce el modelo con los datos de entrenamiento
als_model = als.fit(train)
```

```python
# Se transforman los datos del testeo en base a los datos de entenamiento
predictions = als_model.transform(test)
```

```python
# Remover las predicciones vacias
final_predictions = predictions.filter(predictions.prediction != float('nan'))
```

```python
# Importacion de la libreria para evaluar la el error que puede tener la prediccion
from pyspark.ml.evaluation import RegressionEvaluator
# Le evaluacion se hara con RSME
evaluator_rmse = RegressionEvaluator(metricName="rmse", labelCol="rating", predictionCol="prediction")
```

```python
# SE evaluan los datos de la prediccion en el modelo evaluador
rmse = evaluator_rmse.evaluate(final_predictions)
```

```python
# Impresion del RSME, cual es el error que puede tener el modelo
print(f'RSME: {str(rmse)}')
```

```python
# RAW Predictions
final_predictions.toPandas()
```

```python
# Importacion de la funcion de redondeo para redondear las columnas en valores enteros no decimales
from pyspark.sql.functions import round
format_predictions = final_predictions.withColumn('int_prediction', round("prediction", 0))
```

```python
# Datos formateados
format_predictions.toPandas()
```

```python
# Se describen los valores en los que ronda la prediccion
format_predictions.select('int_prediction').describe().show()
```

```python
# Mostrar Recomendaciones de telefonos para los usuarios
user_recomends = als_model.recommendForAllUsers(5)
user_recomends.show(truncate = False)
```

```python
# Mostrar Recomendaciones de usuarios para los telefonos
item_recommends = als_model.recommendForAllItems(5)
item_recommends.show(truncate = False)
```
